```python
import pandas as pd
df = pd.read_csv('BollywoodMovieDetail.csv')
```

```python
df.head(10)
```

```python
#question-1:
print("Total records:",len(df))

print("\n Metadata:")
print(df.info())
```

```python
#question-2:
genre_counts = df['genre'].value_counts()
print(genre_counts)
print(f"Most released genre: {genre_counts.idxmax()}")
```

```python
#question-3:
cross_tab = pd.crosstab(df['genre'], df['releaseDate'])
print(cross_tab)
```

```python
#question-4:
df['releaseDate'] = pd.to_datetime(df['releaseDate'], errors='coerce')
df['Month'] = df['releaseDate'].dt.month
month_counts = df['Month'].value_counts()
print("Peak release month:", month_counts.idxmax())
```

```python
#question-5:
high_budget = df[df['budget'] >= 25]
high_budget_months = high_budget['Month'].value_counts()
print("Month with most high-budget releases:", high_budget_months.idxmax())

```

```python
#question 6:
df['ROI'] = (df['BoxOfficeCollection'] - df['Budget']) / df['Budget']
top_roi = df[['MovieName', 'ROI']].sort_values(by='ROI', ascending=False).head(10)
print(top_roi)
```

```python
#question-7:
roi_by_release = df.groupby('releaseDate')['ROI'].mean().sort_values(ascending=False)
print("Average ROI by Release Time:\n", roi_by_release)
```

```python
#question-8:
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 5))
sns.histplot(df['budget'], bins=30, kde=False)
plt.title("Histogram of Movie Budgets")
plt.xlabel("Budget (Crores)")
plt.ylabel("Frequency")
plt.show()

sns.displot(df['budget'], kde=True)
```

```python
#question-9:
comedy_roi = df[df['genre'] == 'Comedy']['ROI']
drama_roi = df[df['genre'] == 'Drama']['ROI']

sns.kdeplot(comedy_roi, label='Comedy', shade=True)
sns.kdeplot(drama_roi, label='Drama', shade=True)
plt.title("ROI Distribution: Comedy vs Drama")
plt.legend()
plt.show()
```

```python
#question-10:
correlation = df['BoxOfficeCollection'].corr(df['YoutubeLikes'])
print("Correlation between BoxOffice and YouTube Likes:", correlation)

```

```python
#question-11:
plt.figure(figsize=(12, 6))
sns.boxplot(x='genre', y='YoutubeLikes', data=df)
plt.xticks(rotation=45)
plt.title("YouTube Likes by Genre")
plt.show()
```

```python
#question-12:
selected_cols = ['Budget', 'BoxOfficeCollection', 'YoutubeView', 'YoutubeLikes', 'YoutubeDislikes']
sns.pairplot(df[selected_cols])
plt.show()

correlation_matrix = df[selected_cols].corr()
print("Correlation Matrix:\n", correlation_matrix)

```
